```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
data = pd.read_csv("Team_Sports_Survey_Preprocessed.csv")
```

```python
# Split the values into a single items
data["Expected Shower Room Items (For Swimming Participants)"] = data["Expected Shower Room Items (For Swimming Participants)"].str.split(",")

# Convert the items into a transactional data 
transactions = data["Expected Shower Room Items (For Swimming Participants)"].explode().str.strip().reset_index()
transactions_encoded = transactions.pivot_table(index="index", columns="Expected Shower Room Items (For Swimming Participants)", aggfunc=lambda x: 1, fill_value=0) 
```

```python
# Use Apriori algorithm to find frequent association
frequent_items= apriori(transactions_encoded, min_support=0.5, use_colnames=True) # threshold 50%
rule = association_rules(frequent_items)
rule[['antecedents','consequents','support', 'confidence', 'lift']].sort_values("confidence", ascending=False)
```
